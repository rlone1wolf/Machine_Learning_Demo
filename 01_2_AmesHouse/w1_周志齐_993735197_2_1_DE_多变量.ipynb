# Ames房价数据集上的特征工程

# 简介

Ames房价预测是Kaggle平台上的一个竞赛任务，需要根据房屋的特征来预测亚美尼亚州洛瓦市（Ames，Lowa）的房价。其中房屋的特征x共有79维，响应值y为每个房屋的销售价格（SalePrice）。

Kaggle官网上的任务说明请见：
https://www.kaggle.com/c/house-prices-advanced-regression-techniques

由于房屋的属性较多，且房屋属性类型各异，该数据集有些属性的缺失值也比较多，因此本任务的任务之一是在该数据集上练习特征工程技巧。

Kaggle平台上kernel区有各个竞赛参数者分享的代码，大家可以学习。
下面的代码也参考了很多不同选手分享的代码，在此一并致谢。

### X & X

```python
train_corr = train.corr().abs() #相关性矩阵
train_corr_index = train_corr.index
size = train_corr.shape[1] #for 的范围
threshold = 0.6 #相关性分界点
corr_list = [] 
# 存储格式: [相关性大小, 轴i, 轴j]
for i in range(0, size):
    for j in range(i+1, size):
        corr_list.append([train_corr.iloc[i, j], i, j])  
# 按相关性大小进行排序
corr_list_sorted = sorted(corr_list, key=lambda x:-x[0])
# 输出相关性大于threshold的
for v, i, j in corr_list_sorted:
    if v > threshold:
        print ("%.3f is %s and %s" % (v
                                      , train_corr_index[i]
                                      , train_corr_index[j]))
```

```python
for v, i, j in corr_list_sorted:
    if v > 0.7: #相关系数大于0.7
        temp1 ,temp2 =train_corr_index[i], train_corr_index[j]
        temp = train.loc[:,[temp1,temp2]]
        plt.scatter(temp.loc[:,temp1], temp.loc[:,temp2], alpha=.5,color='b')
        plt.xlabel(temp1)
        plt.ylabel(temp2)
        plt.title('Correlation coefficient is %s'% (v))
#        overlay = 'Correlation coefficient is: {}'.format(v)
#        plt.annotate(s=overlay, xy=(12.1, 10.6), size='x-large')
        plt.show()
```
