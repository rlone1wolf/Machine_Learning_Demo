# eventSim

```python
#eventPropMatrix : start_time, city, state, zip, country, lat, and lng
#eventContMatrix : 101 columns of words count
#非词频特征，采用Person相关系数作为相似度
#对词频特征，采用余弦相似度，也可以用直方图交/Jacard相似度
eventPropSim = sio.mmread(dpath+"EV_eventPropSim")
eventContSim = sio.mmread(dpath+"EV_eventContSim")
print('type  is:',type(eventPropSim), type(eventContSim))
print('shape is:',eventPropSim.shape, eventContSim.shape)
```

```python
print(eventPropSim.getrow(3343))
#getrow(eventIndex[])
#代表该eventIndex[]与哪些eventIndex[]相似
#值代表相似度是多少
```

```python
print(eventContSim.getrow(3343))
#getrow(eventIndex[])
#代表该eventIndex[]与哪些eventIndex[]相似
#值代表相似度是多少
```

```python
print(eventPropSim.getrow(3343).getcol(287))
#getrow(eventIndex[]).getcol(eventIndex[])
#代表该eventIndex[]与另一eventIndex[]相似
#值代表相似度是多少
```

# userMatrix&Sim

```python
#elocale, birthyear, gender, joinedAt, location, timezone
#去掉user_id列, 特征编码且normalize后为 userMatrix
userMatrix = sio.mmread(dpath+"US_userMatrix")
#计算用户相似度矩阵, 对称矩阵, 对角线为1(自己与自己)
#Person相关系数做为相似度度量
userSimMatrix = sio.mmread(dpath+"US_userSimMatrix")
print('type  is:',type(userMatrix), type(userSimMatrix))
print('shape is:',userMatrix.shape, userSimMatrix.shape)
```

```python
print(userMatrix.getrow(123))
#getrow(userIndex[])
#userIndex[]对应的编码后的特征
```

```python
print(userSimMatrix.getrow(123))
#getrow(userIndex[])
#代表该userIndex[]与哪些userIndex[]相似
#值代表相似度是多少
```

```python
print(userSimMatrix.getrow(123).getcol(3161))
#getrow(userIndex[]).getcol(userIndex[])
#代表该userIndex[]与另一userIndex[]相似
#值代表相似度是多少
```
